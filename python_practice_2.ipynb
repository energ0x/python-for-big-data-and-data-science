# VI. Множини (Set).

17. Створити множину без елементів. Після цого за допомогою методу додайте кілька різних елементів до множини. Чи множини є змінним типом даних?

```python
my_set = set()
my_set.add(10)
my_set.add("random text")
my_set.add(0.1)
print(my_set)
```

*місце для текстової відповіді: так множини є змінним типом даних, можна додавати (add, update), видаляти (remove, discard, pop) елементи з множини після її створення

18. Створити множину, маючи список my_list = [1, 1, 2, 67, 67, 8, 9]. Пояснити, чому "зникли" деякі елементи.

```python
my_list = [1, 1, 2, 67, 67, 8, 9]
my_set = set(my_list)
print(my_set)
```

*місце для текстової відповіді: зникло одне число 67, бо в множині тільки унікальні об'єкти можуть міститись. через це дублікати зникають

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
my_set = {1, 2, 3, 4}
my_set_1 = {3, 4, 5, 6}
res_union = my_set.union(my_set_1)
print(res_union)
res_diff = my_set.difference(my_set_1)
print(res_diff)
res_intersection = my_set.intersection(my_set_1)
print(res_intersection)
res_sym_diff = my_set.symmetric_difference(my_set_1)
print(res_sym_diff)
```
