```python
class Cube(Figure_3D):
    def __init__(self, side_length):
        self.side_length = side_length

    def get_volume(self):
        return self.side_length**3

class Cylinder(Figure_3D):
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height

    def get_volume(self):
        return (3.14*self.radius**2)*self.height

cube = Cube(3)
cylinder = Cylinder(2, 5)
print(round((cube+cylinder),1))
```

8. (13–±). –°—Ç–≤–æ—Ä—ñ—Ç—å –∫–ª–∞—Å EWallet –∑ –ø—Ä–∏–≤–∞—Ç–Ω–∏–º–∏ –∞—Ç—Ä–∏–±—É—Ç–∞–º–∏ balance —Ç–∞ id. –†–µ–∞–ª—ñ–∑—É–π—Ç–µ –º–µ—Ç–æ–¥–∏ –ø–æ–ø–æ–≤–Ω–µ–Ω–Ω—è —Ç–∞ –∑–Ω—è—Ç—Ç—è –∫–æ—à—Ç—ñ–≤, –∑–∞–±–µ–∑–ø–µ—á–∏–≤—à–∏ –Ω–∞–ª–µ–∂–Ω—É —ñ–Ω–∫–∞–ø—Å—É–ª—è—Ü—ñ—é. –ü—ñ–¥–∫–∞–∑–∫–∞: –≤–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –¥–µ–∫–æ—Ä–∞—Ç–æ—Ä–∏ property —Ç–∞ setter.

```python
class EWallet:
    def __init__(self, id, balance=0):
        self.__balance = balance
        self.__id = id

    @property
    def balance(self):
        return self.__balance

    @property
    def id(self):
        return self.__id

    @id.setter
    def id(self, new_id):
        self.__id = new_id

    def deposit(self, amount):
        if amount>0:
            self.__balance += amount
            print(f"–ü–æ–ø–æ–≤–Ω–µ–Ω–æ –Ω–∞ {amount}. –ë–∞–ª–∞–Ω—Å: {self.__balance}")
        else:
            print("–ù–µ –∂–ª–æ–±–∏—Å—å")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"–ó–Ω—è—Ç–æ {amount}. –ë–∞–ª–∞–Ω—Å: {self.__balance}")
        else:
            raise ValueError("–ú–∞–ª–æ –≥—Ä–æ—à–µ–π")

wallet_123 = EWallet("user_123", 1000)
print(f"ID: {wallet_123.id}")
print(f"–ë–∞–ª–∞–Ω—Å: {wallet_123.balance}")
wallet_123.deposit(500)
wallet_123.withdraw(200)
```

9. (11–±). –°—Ç–≤–æ—Ä—ñ—Ç—å –∫–ª–∞—Å ShoppingCart, —è–∫–∏–π –º—ñ—Å—Ç–∏—Ç—å —Å–ø–∏—Å–æ–∫ –æ–±'—î–∫—Ç—ñ–≤ —Ç–∏–ø—É Item. –†–µ–∞–ª—ñ–∑—É–π—Ç–µ –º–µ—Ç–æ–¥–∏ –¥–ª—è –¥–æ–¥–∞–≤–∞–Ω–Ω—è –∞–π—Ç–µ–º—ñ–≤, —ó—Ö –≤–∏–¥–∞–ª–µ–Ω–Ω—è —Ç–∞ –≤—ñ–¥–æ–±—Ä–∞–∂–µ–Ω–Ω—è —Å–ø–∏—Å–∫—É –∞–π—Ç–µ–º—ñ–≤.

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f"–î–æ–¥–∞–Ω–æ {item} –¥–æ –∫–æ—à–∏–∫–∞")

    def remove_item(self, item):
        if item in self.items:
            self.items.remove(item)
            print(f"–í–∏–¥–∞–ª–µ–Ω–æ {item} –∑ –∫–æ—à–∏–∫–∞")
        else:
            print(f"{item} –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–æ –≤ –∫–æ—à–∏–∫—É")

    def show_items(self):
        print("–°–ø–∏—Å–æ–∫ —Ç–æ–≤–∞—Ä—ñ–≤ —É –∫–æ—à–∏–∫—É:")
        for item in self.items:
            print(f"- {item}")

# "–º—ñ—Å—Ç–∏—Ç—å —Å–ø–∏—Å–æ–∫ –æ–±'—î–∫—Ç—ñ–≤ —Ç–∏–ø—É Item"
class Item:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"{self.name} ({self.price} –≥—Ä–Ω)"

shopping_cart = ShoppingCart()
phone = Item("Samsung A55", 15000)
laptop = Item("Thinkpad", 30000)
shopping_cart.add_item(phone)
shopping_cart.add_item(laptop)
shopping_cart.show_items()
shopping_cart.remove_item(phone)
shopping_cart.show_items()
```

13. (11–±). –°—Ç–≤–æ—Ä—ñ—Ç—å –∫–ª–∞—Å, —â–æ –º–∞—î –≤–∏–∑–Ω–∞—á–µ–Ω–∏–π Iterator Protocol —Ç–∞ –ø—Ä–∏ –∫–æ–∂–Ω–æ–º—É –≤–∏–∫–ª–∏–∫—É next –ø–æ–≤–µ—Ä—Ç–∞—î –Ω–∞–∑–≤—É –¥–Ω—è —Ç–∏–∂–Ω—è, –ø–æ–∫–∏ –≤–æ–Ω–∏ –Ω–µ –≤–∏—á–µ—Ä–ø–∞—é—Ç—å—Å—è. –ó–≤–µ—Ä–Ω—ñ—Ç—å —É–≤–∞–≥—É, —â–æ –ø–æ—Ä—è–¥–æ–∫ –º–∞—î –∑–Ω–∞—á–µ–Ω–Ω—è.

```python
class Weekday:
    def __iter__(self):
        self.days = ["–ü–æ–Ω–µ–¥—ñ–ª–æ–∫", "–í—ñ–≤—Ç–æ—Ä–æ–∫", "–°–µ—Ä–µ–¥–∞", "–ß–µ—Ç–≤–µ—Ä", "–ü'—è—Ç–Ω–∏—Ü—è", "–°—É–±–æ—Ç–∞", "–ù–µ–¥—ñ–ª—è"]
        self.index = 0
        return self

    def __next__(self):
        if self.index < len(self.days):
            day = self.days[self.index]
            self.index += 1
            return day
        else:
            raise StopIteration

weekdays = Weekday()
for day in weekdays:
    print(day)
```
